## Seaborn - Categorial Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips, estimator = np.std)
```

```python
sns.countplot(x='sex', data = tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill', data=tips, hue = 'smoker')
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter = True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, jitter = True, hue = 'sex')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

## Seaborn - Matrix Plot

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
flight = sns.load_dataset('flights')
tips.head()
```

```python
flight.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot = True)
```

```python
sns.heatmap(tc,annot = True, cmap = 'coolwarm')
```

```python
fp = flight.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap = 'magma')
```

```python
sns.heatmap(fp, cmap = 'magma', linecolor = 'white', linewidths = 1)
```

```python
sns.clustermap(fp)
```

```python
sns.clustermap(fp, cmap = 'coolwarm', standard_scale = 1)
```

## Seaborn - Regression Plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers = ['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='sex', row = 'time')
```

# Grids

```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.barplot)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g =  sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot,'total_bill')
```

```python
import matplotlib.pyplot as plt
g1 =  sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g1.map(plt.scatter,'total_bill','tip')
```

# Style and Color

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_style('ticks')
sns.countplot(x = 'sex', data = tips)
sns.despine(left = True, bottom = True)
```

```python
sns.set_style('ticks')
sns.countplot(x = 'sex', data = tips)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x = 'sex', data = tips)
```

```python
sns.set_context('poster')
sns.countplot(x = 'sex', data = tips)
```
